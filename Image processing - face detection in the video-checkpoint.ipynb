```python
import sklearn.preprocessing as sp
d

encoder=sp.OneHotEncoder()
Y_1=encoder.fit_transform(np.c_[Y]).toarray()
pd.DataFrame(Y_1)
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(X,Y_1,test_size=0.3,random_state=42)
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers


network = models.Sequential()
# network.add(layers.Flatten())
network.add(layers.Dense(1024 , input_dim=90000 , activation = 'relu' ))
network.add(layers.Dense(512  , activation = 'relu' ))
network.add(layers.Dense(256, activation = 'relu' ))

network.add(layers.Dense(3, activation = 'softmax'))

```

```python
network.compile(loss = 'categorical_crossentropy',optimizer = 'adam',metrics=['acc'])
```

```python
network.fit(train_x , train_y , epochs = 60)
```

```python
a=network.predict(test_x)
a
b=encoder.inverse_transform(a)
b
```
